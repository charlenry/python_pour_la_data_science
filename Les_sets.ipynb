# **Les sets (ou les ensembles)**

```python
# Les Sets (les ensembles) sont utilisés pour stocker plusieurs éléments dans une seule variable.
# Set est l’un des 4 types de données intégrés en Python utilisés pour stocker des collections de données, les 3 autres sont List, Tuple et Dictionary, tous avec des qualités et un usage différents.
# Un Set est une collection qui n’est pas ordonnée, immuable* et non indexée. Les Sets sont écrits avec des accolades.
# Les doublons ne sont pas autorisés dans un Set.
# Un Set peut contenir différents types de données.
```

```python
this_set = {"apple", "banana", "cherry"}
print(this_set)
```

```python
# Obtenir la longueur d'un Set
print(len(this_set))
```

```python
# Accéder à un élément d'un Set
# Vous ne pouvez pas accéder aux éléments d'un Set en vous référant à un index, car les Sets ne sont pas indexés.
# Mais vous pouvez boucler à travers les éléments d'un Set en utilisant une boucle for, ou demander si un élément spécifique est présent dans un Set, en utilisant le mot clé in.
for x in this_set:
    print(x)
```

```python
# Vérifiez si « banana » est présent dans le Set
print("banana" in this_set)
```

```python
# Vérifiez si « banana » n’est PAS présent dans le Set
print("banana" not in this_set)
```

```python
  # Ajoutez un élément à un Set à l’aide de la méthode add()
this_set.add("orange")
print(this_set)
```

```python
# Ajoutez plusieurs éléments à un Set à l’aide de la méthode update()
this_set.update(["peach", "mango", "grapes"])
print(this_set)
```

```python
# Ajouter une liste à un Set
mylist = ["kiwi", "pineapple"]
this_set.update(mylist)
print(this_set)
```

```python
# Supprimer un élément d'un Set à l’aide de la méthode remove()
# Si l'élément à supprimer n'existe pas, remove() générera une erreur.
this_set.remove("banana")
print(this_set)
```

```python
# Supprimer un élément d'un Set à l’aide de la méthode discard()
# Si l'élément à supprimer n'existe pas, discard() ne générera pas d'erreur.
this_set.discard("pineapple")
```

```python
# Supprimez un élément aléatoire à l’aide de la méthode suivante :pop()
# Les Sets sont non ordonnés, vous ne saurez donc pas quel élément sera supprimé.
x = this_set.pop()
print(x, "a été retiré du Set")
print(this_set)
```

```python
# Supprimez tous les éléments d'un Set à l’aide de la méthode clear()
this_set.clear()
print(this_set)
```

```python
# Supprimez le Set complet à l’aide de la méthode del
# del this_set
# print(this_set) # Cela générera une erreur car le Set n'existe plus.
```

```python
# Joignez deux Sets en utilisant la méthode union()
set1 = {"a", "b", "c"}
set2 = {1, 2, 3}
new_set = set1.union(set2)
# ou en utilisant l'opérateur | entre les deux Sets :
# set3 = set1 | set2
print(new_set)
```

```python
set3 = {"John", "Elena"}
set4 = {"apple", "bananas", "cherry"}

# Joignez les set1, set2, set3 et set4 en un seul Set 
new_set = set1.union(set2, set3, set4)
print(new_set)
```

```python
# Joignez set1 et set2, mais ne conservez que les doublons en utilisant la méthode intersection() ou l'opérateur & :

set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

new_set = set1.intersection(set2)
# ou en utilisant l'opérateur & entre les deux Sets :
# new_set = set1 & set2
print(new_set)
```

```python
# Conservez tous les éléments de set1 qui ne sont pas dans set2 en utilisant la méthode difference() ou l'opérateur - :
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

new_set = set1.difference(set2)
# ou en utilisant l'opérateur - entre les deux Sets :
# new_set = set1 - set2
print(new_set)
```

```python
# Conservez les articles qui ne sont pas présents dans les deux ensembles en utilisant la méthode symmetric_difference() ou l'opérateur ^:
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

new_set = set1.symmetric_difference(set2)
# ou en utilisant l'opérateur ^ entre les deux Sets :
# new_set = set1 ^ set2

print(new_set)
```
