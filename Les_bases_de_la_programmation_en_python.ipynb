## <font color="lime">Problème à résoudre de la capsule 25</font>

**Objectif : Manipuler une liste contenant les prix de 58 maisons**
1. Créer une liste nommée « prix_de_58_maisons » contenant des prix entre 125000 et 700000 euros
2. Combien de maisons ont un prix supérieur ou égal à 300000 euros ?
3. Combien de maisons ont un prix compris entre 250000 et 400000 euros ?
4. Combien de maisons ont un prix qui n’est pas supérieur à 600000 euros ?
5. Combien de maisons ont un prix inférieur à 150000 euros ou supérieur à 650000 euros ?

```python
# 1. Créer une liste nommée « prix_de_58_maisons » contenant des prix entre 125000 et 700000 euros
prix_de_58_maisons=list(range(125000, 700000, 10000))
print(prix_de_58_maisons)
len(prix_de_58_maisons)
```

```python
# 2. Combien de maisons ont un prix supérieur ou égal à 300000 euros ?
compteur = 0
for prix in prix_de_58_maisons:
    if prix >= 300000:
        compteur += 1
print("Il y a %d maisons qui ont un prix supérieur ou égal à 300000 euros" % compteur)
```

```python
# 3. Combien de maisons ont un prix compris entre 250000 et 400000 euros ?
compteur = 0
for prix in prix_de_58_maisons:
    if prix >= 250000 and prix <= 400000:
        compteur += 1
print("Il y a %d maisons qui ont un prix compris entre 250000 et 400000 euros" % compteur)
```

```python
# 4. Combien de maisons ont un prix qui est inférieur ou égal à 600000 euros ?
compteur = 0
for prix in prix_de_58_maisons:
    if prix <= 600000:
        compteur += 1
print("Il y a %d maisons qui ont un prix inférieur ou égal à 600000 euros" % compteur)
```

```python
# 5. Combien de maisons ont un prix inférieur à 150000 euros ou supérieur à 650000 euros ?
compteur = 0
for prix in prix_de_58_maisons:
    if prix < 150000 or prix > 650000:
        compteur += 1
print("Il y a %d maisons qui ont un prix inférieur à 150000 euros ou supérieur à 650000 euros" % compteur)
```

# Les fonctions en Python

```python
# def fonction(parametre1, parametre2, ..., parametreN):
#    instruction
#    return resultat
```

```python
def nombre_de_maisons_avec_prix_superieur_ou_egal_a_300000(une_liste):
    compteur = 0
    for prix in une_liste:
        if prix >= 300000:
            compteur += 1
    return compteur
```

```python
# appel de la fonction nombre_de_maisons_avec_prix_superieur_ou_egal_a_300000
nombre_de_maisons_avec_prix_superieur_ou_egal_a_300000(prix_de_58_maisons)
```

```python
# fonction qui retourne plsieurs résultats
def nombre_de_maisons_avec_prix_superieur_ou_egal_a_300000_et_nombre_de_maisons(une_liste):
    compteur = 0
    nombre_de_maisons = len(une_liste)
    for prix in une_liste:
        if prix >= 300000:
            compteur += 1
    return compteur, nombre_de_maisons
```

```python
# appel de la fonction nombre_de_maisons_avec_prix_superieur_ou_egal_a_300000_et_nombre_de_maisons
(resultat, total) = nombre_de_maisons_avec_prix_superieur_ou_egal_a_300000_et_nombre_de_maisons(prix_de_58_maisons)
print("Il y a %d maisons qui ont un prix supérieur ou égal à 300000 euros sur un total de %d maisons" % (resultat, total))
```

```python
# Fonction lambda
# lambda parametre1, parametre2, ..., parametreN: expression
# lambda x, y: x + y
# lambda x, y: x * y

# to call a lambda function, you need to assign it to a variable
# addition = lambda x, y: x + y
# multiplication = lambda x, y: x * y

# addition(2, 3)
# multiplication(2, 3)
```

```python
# lambda without arguments
greet = lambda : print('Hello World')

# lambda call
greet()
```

```python
# lambda that accepts one argument
greet_user = lambda name : print('Hey there,', name)

# lambda call
greet_user('Delilah')
```
